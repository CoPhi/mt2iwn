```python
import xml.etree.ElementTree as ET
import csv
```

```python
# File paths for MariTerm and ItalWordnet files
MariT_file = '/content/drive/MyDrive/MariT2ItalWN/MariT_03_24.xml'
IWN_file = '/content/drive/MyDrive/MariT2ItalWN/IWN_03_24.xml'
lemmas_csv = '/content/drive/MyDrive/MariT2ItalWN/candidates_06_24.csv'
```

```python

# Function to extract literal lemmas and their details from an XML file
def extract_lemmas(file_path):
    tree = ET.parse(file_path)
    root = tree.getroot()
    lemmas_with_details = {}

    for word_meaning in root.findall('.//WORD_MEANING'):
        word_meaning_id = word_meaning.get('ID')
        literal = word_meaning.find('.//VARIANTS//LITERAL')
        literal_lemma = literal.get('LEMMA') if literal is not None else None
        if literal_lemma:
            lemmas_with_details[literal_lemma] = word_meaning_id

    return lemmas_with_details

# Function to find shared literal lemmas between two resources
def find_shared(MariT_file, IWN_file):
    MariT_lemmas = extract_lemmas(MariT_file)
    IWN_lemmas = extract_lemmas(IWN_file)

    shared_lemmas = set(MariT_lemmas.keys()).intersection(set(IWN_lemmas.keys()))
    print(f"Total extracted lemmas: {len(shared_lemmas)}")

    return shared_lemmas, MariT_lemmas, IWN_lemmas

# Function to write shared lemmas to a CSV
def write_csv(shared_lemmas, MariT_lemmas, IWN_lemmas, file_name):
    with open(file_name, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file, delimiter=';')
        writer.writerow(['Shared Lemma', 'IWN_WM_ID', 'MarT_WM_ID'])
        for lemma in sorted(shared_lemmas, key=lambda x: x.lower()):
            writer.writerow([lemma, IWN_lemmas[lemma], MariT_lemmas[lemma]])
```

```python
# Step 1: Find shared literal lemmas between the two resources
shared_lemmas, MariT_lemmas, IWN_lemmas = find_shared(MariT_file, IWN_file)

# Step 2: Write shared lemmas to CSV

write_csv(shared_lemmas, MariT_lemmas, IWN_lemmas, lemmas_csv)

print(f"Shared lemmas written to {lemmas_csv}")
```
