```python
from collections import defaultdict
import os
import xml.etree.ElementTree as ET
```

```python

# Function definition
def count_xml_data(xml_file):
    # Parse the XML file
    tree = ET.parse(xml_file)
    root = tree.getroot()

    # Initialize counters
    word_meaning_count = 0
    unique_lemmas = set()
    synset_relation_count = 0

    # Iterate over <WORD_MEANING> nodes
    for word_meaning in root.findall(".//WORD_MEANING"):
        word_meaning_count += 1

        # Find all <LITERAL> nodes in <VARIANTS>
        for literal in word_meaning.findall(".//LITERAL"):
            lemma = literal.attrib.get("LEMMA")
            sense = literal.attrib.get("SENSE")

            if lemma:
                unique_lemmas.add(lemma)


        # Count relations in <INTERNAL_LINKS>
        internal_links = word_meaning.find("INTERNAL_LINKS")
        if internal_links is not None:
            synset_relation_count += len(internal_links.findall("RELATION"))

    # Prepare the results
    results = {
        "Word Meanings": word_meaning_count,
        "Unique Lemmas": len(unique_lemmas),
        "Synset Relations": synset_relation_count
    }

    return results



# Store results for each XML file
all_results = []

# Iterate through each XML file in the directory
for xml_file in os.listdir(xml_directory):
    if xml_file.endswith('.xml'):  # Check if the file is an XML
        file_path = os.path.join(xml_directory, xml_file)
        result = count_xml_data(file_path)
        result['File Name'] = xml_file  # Add filename to results
        all_results.append(result)

# Print results for each XML file
print("Results Breakdown for each XML file:")
for res in all_results:
    print(res)
```

```python
duplicates = []

def find_duplicates_in_synsets(file_path):
    """
    Finds duplicate synsets in the XML file based on the same ID and same first literal lemma.

    Args:
        file_path (str): Path to the XML file.

    Returns:
        None: Prints the number of duplicates and detailed information.
    """
    tree = ET.parse(file_path)
    root = tree.getroot()

    # Dictionary to track synsets by ID
    synset_dict = defaultdict(list)


    # Iterate over all WORD_MEANING elements
    for word_meaning in root.findall(".//WORD_MEANING"):
        synset_id = word_meaning.attrib.get("ID", "unknown")
        literals = word_meaning.findall(".//VARIANTS/LITERAL")

        # Get the literal lemmas in the order they appear
        literal_lemmas = [literal.attrib.get("LEMMA", "unknown") for literal in literals]

        # If there are no literals, continue to the next element
        if not literal_lemmas:
            continue

        first_literal = literal_lemmas[0]  # First literal lemma

        # Check for duplicates based on ID and first literal
        key = (synset_id, first_literal)
        if key in synset_dict:
            duplicates.append((key, synset_dict[key], literal_lemmas))
        else:
            synset_dict[key] = literal_lemmas

    # Print results
    print(f"Total duplicates found: {len(duplicates)}")
    if duplicates:
        print("\nDuplicate details:")
        for (synset_id, first_literal), original_literals, duplicate_literals in duplicates:
            print(
                f"ID: {synset_id}; {', '.join(original_literals)} | "
                f"ID: {synset_id}; {', '.join(duplicate_literals)}"
            )

find_duplicates_in_synsets(IWN_pre_mod)
```
